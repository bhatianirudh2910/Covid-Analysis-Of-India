```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s = pd.read_csv('countries-aggregated.csv',encoding='utf-8').fillna(0)
```

```python
r = s['Recovered'].iloc[0:320].values
d = s['Deaths'].iloc[0:320].values
a = s['Active'].iloc[0:320].values
date = s['Date']
```

```python
def active():
        fig = plt.figure(figsize=(15,7))
        ax = fig.add_subplot(111)
        ax.set_title("Active cases")
        ax.set_xlabel('date')
        ax.set_ylabel('Active cases')
        ax.plot(date,a,label='Active')
        leg = ax.legend()
        plt.show()

def recovered():
        fig = plt.figure(figsize=(15,7))
        ax = fig.add_subplot(111)
        ax.set_title("Recovered")
        ax.set_xlabel('date')
        ax.set_ylabel('Recovered')
        ax.plot(date,r,'g',label='Recovered')
        leg = ax.legend()
        plt.show()
        
def deaths():
        fig = plt.figure(figsize=(15,7))
        ax = fig.add_subplot(111)
        ax.set_title("Deaths")
        ax.set_xlabel('date')
        ax.set_ylabel('Deaths')
        ax.plot(date,d,'r',label='Deaths')
        leg = ax.legend()
        plt.show()
        
```

```python
active()
```
